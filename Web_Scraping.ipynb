```
from bs4 import BeautifulSoup
```

```
!wget -O simple.html https://www.dropbox.com/s/8mva10vrgpvh5cq/simple.html?dl=0
```

```
with open('simple.html') as html_file :
    soup = BeautifulSoup(html_file,'lxml')      
```

```
print(soup.prettify())
```

```
match=soup.title.text
print(match)
```

```
match=soup.div
print(match)
```

```
match=soup.find('div',class_='footer')
print(match)
```

```
match=soup.find('div',class_='footer').text
print(match)
```

```
article=soup.find('div',class_='article').text
print(article)
```

```
#Code bit to grab div class article  in that h2 in that the areticle text 
article=soup.find('div',class_='article').h2.a.text
print(article)
```

```
#Grab para of the body 
article=soup.find('div', class_='article').p.text
print(article)
```

```
#Now we can get a single article header and its body 
#Now how to loop through a list of articles 


for article in soup.find_all('div',class_= 'article') : 
    headline = article.h2.a.text
    print(headline)

    summary = article.p.text
    print(summary)

    print()
```
