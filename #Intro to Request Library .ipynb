```
#get parameters 
 payload = {'page':2,'count':25}
 r = requests.get('https://httpbin.org/get',params= payload)

 print(r.text)
```

```
print(r.url)
```

```
#Sometime you may want to post some data into the site 
payload = {'username':'corey','password':'testing'}
r = requests.post('https://httpbin.org/post',data= payload)

print(r.text)
```

```
# the above is JSON responce 
# instead you can use a JSON method instead 

print(r.json())
```
